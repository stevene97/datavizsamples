```python
# Import packages
import pandas as pd
import numpy as np
import geocoder
import folium
import folium.plugins
from folium.plugins import MarkerCluster
```

```python
# Read .csv file with pandas
df = pd.read_csv('/Users/steven_esp12/Desktop/reclamosfinalpermonth/reclamosfinaljunio.csv')

# First five entries
df.head()
```

```python
# Creating list of coordinates
coordinates = []

for i in range(0, len(df)):
    g = geocoder.arcgis(df['Direccion'][i])
    coordinates.append(g.latlng)
```

```python
# Adding coordinates to DataFrame
df['Coordinates'] = coordinates

# Save new DataFrame
df.to_csv('/Users/steven_esp12/Desktop/reclamosfinalpermonth/reclamosfinaljunio2.csv', index=0)

# Uploading new DataFrame
df2 = pd.read_csv('/Users/steven_esp12/Desktop/reclamosfinalpermonth/reclamosfinaljunio2.csv')

# Check new Dataframe
df2.head()
```

```python
#-------------------------CLEANING OBSERVATIONS-------------------------
# Dropping observations with NaN
df3 = df2.dropna(how='any')

# Check observations were dropped.
df3.head()
```

```python
# Creating popups
popups = []

for index, row in df3.iterrows():
    code = '<b>Fecha: </b>' + row['Fecha Creacion'] + '<br><b>Dirección: </b>' + row['Direccion'].replace("\'", " ") + '<br><b>Grupo Planificador: </b>' + row['Grupo Planificador'] + '<br><b>Texto Pres.: </b>' + row['Texto Pres.'] + '<br><a href="http://solicitudes.buenosaires.gob.ar/suaci/contacto">Ingresá un reclamo</a>' + ' | <a href="https://www.facebook.com/GCBA/posts/ingres%C3%A1-tus-reclamos-denuncias-quejas/10151044266348613/">Pagina de Facebook</a>'
    popups.append(code)
```

```python
# Creating new list of coordinates
coordinatesdf3 = []

for index, row in df3.iterrows():
    coordinatesdf3.append(eval(row['Coordinates']))
```

```python
# Ensuring equal lengths (otherwise distorted information)
len(popups) == len(coordinatesdf3)
```

```python
# New Dataframe
dataframei = pd.DataFrame({"Coordinates":coordinatesdf3, "Popup screen":popups})
```

```python
#--------------DROPPING INCORRECT OBSERVATIONS-------------------
malindex = []

for index, row in dataframei.iterrows():
    if list(dataframei['Coordinates'][index])[0] > -34.522308 or list(dataframei['Coordinates'][index])[0] < -34.707248 or list(dataframei['Coordinates'][index])[1] < -58.531782 or list(dataframei['Coordinates'][index])[1] > -58.329309:
        malindex.append(index)
```

```python
malindex
```

```python
dataframei.drop(dataframei.index[[22, 308, 331, 351, 446, 459, 472, 529]], inplace=True)
```

```python
# Converting each column into list
dfiloc = list(dataframei['Coordinates'])

dfipopup = list(dataframei['Popup screen'])
```

```python
# Testing equal lengths
len(dfiloc) == len(dfipopup)
```

```python
# Create a map of Buenos Aires with the "OpenStreetMap" layer and a zoom start of 11.5.
m_bsas = folium.Map(location = [-34.61, -58.425], 
                    zoom_start = 11.5, 
                    tiles = "OpenStreetMap", 
                    prefer_canvas = True, 
                    min_zoom = 10,
                    width = 900,
                    height = 500,
                    control_scale = True)

# Adding markers with cluster data
marker_cluster = MarkerCluster().add_to(m_bsas)

for i in range(0, len(dfipopup)):
    folium.Marker(location = dfiloc[i],
                  popup = dfipopup[i]).add_to(marker_cluster)

# creating a fullscreen button
# fullscreen
folium.plugins.Fullscreen(position='topright',
                          force_separate_button = True).add_to(m_bsas)
```
